# Packaged

```python
import pandas as pd
from pathlib import Path
import os
from sklearn.model_selection import train_test_split
from pycaret.classification import *
import pickle

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 120)
```

```python
PATH_REPO = Path(os.getcwd()).parent
PATH_DATA = PATH_REPO / "data/input"
GWG = "GWG"
```

# Read Data

```python
with open(PATH_REPO / "data/models/numeric_features_names.pkl", 'rb') as f:
    numeric_features_names = pickle.load(f)
with open(PATH_REPO / "data/models/categoric_features_names.pkl", 'rb') as f:
    categoric_features_names = pickle.load(f)
with open(PATH_REPO / "data/models/targets_names.pkl", 'rb') as f:
    targets_names = pickle.load(f)
print("Total columns names: Numerics ({}) + Categoric ({}) + Targets ({}) = {}".format(len(numeric_features_names), len(categoric_features_names), len(targets_names), 
                                                                                       len(numeric_features_names)+ len(categoric_features_names)+ len(targets_names) ))
```

```python
data_dict = {}
for t in targets_names:
    print(t)
    path = PATH_DATA / "01_preprocess_balanced_{}.csv".format(t)
    data = pd.read_csv(path, index_col=0)
    data_dict[t] = data
    display(data)
```

# Features Selection

# Compare Models

```python
def fs_cleaner(feat, ex_feat_list):
    for f in ex_feat_list:
        if f in feat:
            return True, f
    return False, feat
```

```python
bypass_feat = [GWG]
data_fs_dict = {}
for t in targets_names:
    print("\n >> TARGET: {} \n".format(t))
    data = data_dict[t]
    fs_model = setup(data = data, target = t, feature_selection = True, remove_multicollinearity = True, feature_selection_threshold=.3, ignore_low_variance = True,
                     silent=True, verbose=False, categorical_features=categoric_features_names, numeric_features=numeric_features_names)
    
    # Find new features list
    for elem in fs_model:
        if isinstance(elem, pd.DataFrame):
            if len(elem.columns) < len(data.columns):
                fs_features_names = elem.columns.tolist()
                break
    
    # Check and add if bypass feat are in features
    for bpf in bypass_feat:
        if bpf not in fs_features_names:
            print("adding bypass feat", bpf)
            fs_features_names.append(bpf)

    fs_numeric_features_names = []
    fs_categoric_features_names = []
    print("Total features:", len(fs_features_names))
    for f in fs_features_names:

        is_check, f_name = fs_cleaner(f, numeric_features_names)

        if is_check:
            fs_numeric_features_names.append(f_name)

        is_check, f_name = fs_cleaner(f, categoric_features_names)

        if is_check:
            fs_categoric_features_names.append(f_name)
    
    data_fs = data[[t] + fs_numeric_features_names + fs_categoric_features_names]
    
    data_fs_dict[t] = {"data": data_fs, "numeric_features_names":  fs_numeric_features_names, "categoric_features_names": fs_categoric_features_names}
    
    display(data_fs)
    
    del fs_model
```

# Benchmark

```python
path_main = PATH_REPO / "data"
best_models_dict = {}
exclude=['knn']
#models = ["lr", "knn"]
for t in targets_names:
    print("\n >> TARGET: {} \n".format(t))
    
    # Train & optimize models
    
    data = data_fs_dict[t]["data"]
    categoric_features_names = data_fs_dict[t]["categoric_features_names"]
    numeric_features_names = data_fs_dict[t]["numeric_features_names"]
    clf = setup(data = data, target = t, train_size=0.7 ,preprocess=False,
                verbose=False, silent=True, categorical_features=categoric_features_names, numeric_features=numeric_features_names)
    best = compare_models(sort = 'AUC', exclude=exclude)
    best_models_dict[t] = {"best": best, "setup": clf, "scoring_grid": pull()}
    
    # plots
    plots_names = ['confusion_matrix', 'feature_all', 'auc']
    for pn in plots_names:
        path = path_main / "plots/{}/{}".format(t, pn)
        plot_model(best, plot = pn, save=path)
        print("file saved in", path)
        
    # prepare output
    
    path = path_main / "models/{}/best_model.pkl".format(t)
    with open(path, "wb") as of:
        pickle.dump(best, of)
        print("file saved in", path)
    
    path = path_main / "models/{}/categoric_features_names.pkl".format(t)
    with open(path, "wb") as of:
        pickle.dump(categoric_features_names, of)
        print("file saved in", path)
    
    path = path_main / "models/{}/numeric_features_names.pkl".format(t)
    with open(path, "wb") as of:
        pickle.dump(numeric_features_names, of)
        print("file saved in", path)    
    
    # output
    
    scoring_grid = pull()
    
    path = path_main / "output/{}/scoring_grid.xlsx".format(t)
    scoring_grid.to_excel(path)
    print("file saved in", path)
    
    
    display(best_models_dict[t]['scoring_grid'])
    
```

# Generate test prediction data

```python
cons_features = []
counter = 0
for t in targets_names:
    print("\n >> TARGET: {} \n".format(t))
    model = best_models_dict[t]["best"]
    categoric_features_names = data_fs_dict[t]["categoric_features_names"]
    numeric_features_names = data_fs_dict[t]["numeric_features_names"]
    features = numeric_features_names + categoric_features_names 
    
    for f in features:
        if f not in cons_features:
            counter += 1
            print("({}) {}".format(counter, f))
            cons_features.append(f)
    
    data = data_dict[t][features].sample(5)
    display(data)
    pred = model.predict_proba(data)
    print(pred)
    
```
